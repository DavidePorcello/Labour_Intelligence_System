```python
storag =[]
for y in range(1,8):
    success = False
    while not success:
        try:
            driver.find_element_by_xpath('/html/body/div[1]/md-container/calc/md-content/div[2]/div[2]/h3').click()
            driver.find_element_by_xpath('/html/body/div[1]/md-container/calc/md-content/div[2]/div[2]/md-content[1]/div/md-radio-group/md-radio-button[{}]'.format(y)).click()
            year = driver.find_element_by_xpath('/html/body/div[1]/md-container/calc/md-content/div[2]/div[2]/md-content[1]/div/md-radio-group/md-radio-button[{}]'.format(y)).text
            success = True
        except:
            pass
    year = driver.find_element_by_xpath('/html/body/div[1]/md-container/calc/md-content/div[2]/div[2]/md-content[1]/div/md-radio-group/md-radio-button[{}]'.format(y)).text
    for net in range(1,60):
        value = driver.find_element_by_xpath('/html/body/div[1]/md-container/calc/md-content/div[1]/md-input-container[1]/input')
        value.clear()
        value.send_keys('{}'.format(net))
        gross = driver.find_element_by_xpath('/html/body/div[1]/md-container/calc/md-content/div[2]/div[1]/div[21]/div[2]').text
        gross = float(''.join(dig for dig in gross if dig.isdigit()))
        storag.append((net,year,gross))
```
